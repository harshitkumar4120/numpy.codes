```python
import numpy as np
import matplotlib.pyplot as plt



exam_data=np.array([
            [1, 80, 98, 95, 93,78],   ## harshit
            [2, 66,	89,	94, 93,69],   ## annu
            [4, 79,	80,	80,	89,90],   ## naina
            [5,	85,	70,	65,	90,90],	  ## akash
            [3,	60,	86,	86,	89,90]    ## anomol
            ])

shape=np.shape(exam_data)
print(shape)
print("\nthe size is:",np.size(exam_data))

print("\n1st three studens data",exam_data[0:3])
total=np.sum(exam_data[:, 1:])
print(total)

```

## total marks by students

```python
import pandas as pd
import matplotlib.pyplot as plt
total=np.sum(exam_data[:, 1:], axis=1)             # axix=1  perform opration only on row and axix=0 on columns
print(total)
```

# some operations

```python
import numpy as np
import matplotlib.pyplot as plt

# for minimum numbers
minimum_num=np.min(exam_data[:, 1:], axis=1)
print(minimum_num)

# for maximum
max_num=np.max(exam_data[:, 1:], axis=1)
print(max_num)

# for average 
average=np.mean(exam_data[:, 1:], axis=1)
print(average)

# for cumsum data
cumsum=np.cumsum(exam_data[:, 1:], axis=1)
print(cumsum)

plt.figure(figsize=(6,5))
plt.plot(np.mean(cumsum, axis=0))
plt.title("average numbers across all students")
plt.xlabel("students")
plt.ylabel("numbers")
plt.grid()
plt.show()
```

## vector operations

```python
vector1=np.array([1,2,3,4,5])
vector2=np.array([2,2,2,2,2])

print("addition:",vector1+vector2)

print("subtraction:",vector1-vector2)

print("product:",vector1*vector2) #cross product.

print("dot product:",np.dot(vector1,vector2))  # dot producd.

```

## vectories operation

```python
student_types=np.array(['Harshit','Annu','Naina','Akash','Anmol'])
vectorized_lower=np.vectorize(str.lower)
print("vectorized:",vectorized_lower(student_types))

student_types=np.array(['harshit','annu','naina','akash','anmol'])
vectorized_upper=np.vectorize(str.upper)
print("vectorized:",vectorized_upper(student_types))
```
